```python
import random

class Account:
    # 클래스 변수 설정: 은행 이름과 계좌 개수
    bank_name = "SC은행"

    #. Q2. 작성
    account_count = 0

    ## Q1.작성
    def __init__(self, owner, initial_balance):
        # 예금주와 초기 잔액 설정
        self.owner = owner
        self.balance = initial_balance
        # 계좌번호는 랜덤하게 생성
        self.account_number = self.generate_account_number()
        # 입금 횟수 추적 변수 초기화
        self.deposit_count = 0
        # Q2. 작성 - 계좌 객체가 생성될 때마다 account_count를 증가
        Account.account_count += 1
        # Q10. 작성 - 입출금내역
        self.deposit_history_list = []
        self.withdraw_history_list = []


    # 계좌번호 생성 메서드
    def generate_account_number(self):
        # 3자리-2자리-6자리 형식의 계좌번호 생성
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"

    # Q3. 작성 - 클래스 메서드: 현재까지 생성된 계좌 수 반환
    @classmethod
    def get_account_num(cls):
        return cls.account_count


    # Q4. 작성 - 입금 메서드
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상이어야 합니다.")
        else:
            self.balance += amount
            self.deposit_count += 1
            print(f"{amount}원이 입금되었습니다. 현재 잔액: {format(self.balance, ',')}원")
            self.deposit_history_list.append(self.balance) # 입금 내역 기록
            # 입금 횟수가 5회일 때 이자 지급
            if self.deposit_count % 5 == 0:
                self.add_interest()
                self.deposit_count = 0 # 5회차 입금 시 , 입금 횟수 초기화
                self.deposit_history_list.append(self.balance) # 입금 내역 기록


    # Q5. 작성 - 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족하여 출금할 수 없습니다.")
        else:
            self.balance -= amount
            print(f"{amount}원이 출금되었습니다. 현재 잔액: {format(self.balance, ',')}원")
            self.withdraw_history_list.append(self.balance) # 출금 내역 기록

    # Q6. 작성 - 정보 출력
    def display_info(self) :
        return f"은행 이름: {self.bank_name} 예금주: {self.owner} 계좌번호: {self.account_number} 잔액: {format(self.balance, ',')}원"


    # Q7.이자 지급 메서드
    def add_interest(self):
        interest = self.balance * 0.01  # 1% 이자 계산
        self.balance += interest
        print(f"이자 {interest}원이 지급되었습니다. 현재 잔액: {format(self.balance, ',')}원")

    ## Q.10 입금, 출금 내역 출력 작성
    # 입금 내역 출력 메서드
    def deposit_history(self):
        print("입금 내역:")
        for idx, amount in enumerate(self.deposit_history_list, 1):
            print(f"{idx}. {amount}원")

    # 출금 내역 출력 메서드
    def withdraw_history(self):
        print("출금 내역:")
        for idx, amount in enumerate(self.withdraw_history_list, 1):
            print(f"{idx}. {amount}원")


    # 객체 정보 출력 메서드
    def __str__(self):
        return f"은행 이름: {self.bank_name}\n예금주: {self.owner}\n계좌번호: {self.account_number}\n잔액: {format(self.balance,',')}원"

############################################################################################################################################
#### 실행
############################################################################################################################################

# Q1. 확인
print('-'*60)
print("Q1. 확인")
q1 = Account("이은정", 70000000)

# Q.3. 확인
print('-'*60)
print("Q3. 확인")
q3 = Account("이덕만", 30000000)
print( f"총 생성된 계좌 수: {q3.get_account_num()}"   )

# Q.4. 확인, Q6. 확인
print('-'*60)
print("Q.4. 확인, Q6. 확인")
q1.deposit(800)
print(  q1.display_info()   )

# Q.5. 확인, Q6. 확인
print('-'*60)
print("Q.5. 확인, Q6. 확인")
q1.withdraw(900)
print(  q1.display_info()   )

# Q.7. 확인, Q6. 확인
print('-'*60)
print("Q.7. 확인, Q6. 확인")
q1.deposit(700)
q1.deposit(600)
q1.deposit(500)
q1.deposit(400)
print(  q1.display_info()   )

q1.deposit(800)
q1.deposit(700)
q1.deposit(600)
q1.deposit(500)
q1.deposit(400)
print(  q1.display_info()   )

# Q.8. 확인
print('-'*60)
print("Q.8. 확인")
q8 = Account("이윤석", 50000000)
# 생성된 인스턴스를 리스트에 저장
accounts = [q1, q3, q8]

# Q.9 확인
print('-'*60)
# 잔고가 100만원 이상인 고객의 정보만 출력
print("Q.9. 확인")
print("---------- 잔고가 100만원 이상인 고객 정보 ----------")
for account in accounts:
    if account.balance >= 1000000:
        print(account)
        print("-" * 53)

print("Q.10. 확인")
print('-'*60)
# Q.10 확인 입금 내역 출력
print("------------- q1 인스턴스의 입출금내역 -------------")
print(q1.display_info())
q1.deposit_history()
q1.withdraw_history()

print("------------- q3 인스턴스의 입출금내역 -------------")
print(q3.display_info())
q3.deposit_history()
q3.withdraw_history()

print("------------- q8 인스턴스의 입출금내역 -------------")
print(q8.display_info())
q8.deposit_history()
q8.withdraw_history()


print( f"총 생성된 계좌 수: {q3.get_account_num()}"   )

```
